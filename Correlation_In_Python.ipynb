```python
#import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

#read the data

df=pd.read_csv(r'C:\Users\SweetSoumyaKavya\Desktop\Bigata\movies\movies.csv')
```

```python
#look at the data

df.head()
```

```python
# check for missing data and drop
for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
df=df.dropna()
df.head()
```

```python
#check data types
df.dtypes
```

```python
# change datatype of columns
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
df['votes']=df['votes'].astype('int64')
#sort the data
df=df.sort_values(by='gross',ascending=False,inplace=False)
df.head()
```

```python
#see all data by scrolling
pd.set_option('display.max_rows',None)
```

```python
df.drop_duplicates().head()
```

```python
#finding correlation

#comparing budget and gross

#scatter plot with budget and gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for film')
plt.ylabel('Gross earnings')
plt.show()
```

```python
df.head()
```

```python
#using seaborn to plot and compare budget and gross

sns.regplot(x='budget',y='gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
#correlation matrix for numerical value
df.corr(method='pearson')  #pearson,kendall,spearman
```

```python
df=df.drop(columns=['year']) #year is not required for finding correlation when gross is the predictor
corr_matrix=df.corr(method='pearson')

plt.figure(figsize=(10,5))
sns.heatmap(corr_matrix, annot=True)

plt.title('Correlation matrix')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
```

```python
#Numerizing all columns

df_numerized = df
for col_name in df_numerized:
    if(df_numerized[col_name].dtype== 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
corr_matrix=df_numerized.corr(method='pearson')

plt.figure(figsize=(15,10))
sns.heatmap(corr_matrix, annot=True)

plt.title('Correlation matrix')
plt.xlabel('Movie features')
plt.ylabel('Movie features')
```

```python
correlation_mat = df_numerized.corr()
correlation_pairs=correlation_mat.unstack()
correlation_pairs
```

```python
sort_pair=correlation_pairs.sort_values()
sort_pair
```

```python
high_corr=sort_pair[(sort_pair)>0.5]
high_corr
```
